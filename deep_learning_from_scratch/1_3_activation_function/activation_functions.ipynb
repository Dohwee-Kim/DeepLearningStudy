## 시그모이드를 구현해 봅시다 !!

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
x = np.arange(-5.0, 5.0 , 0.1)
```

```python
y = sigmoid(x)
```

```python
plt.plot(x,y)
plt.ylim(-0.1, 1.1) #y축 범위 지정 
plt.show()
```

## 자 드디어 Relu 차례입니다

```python
def ReLU(x):
    return np.maximum(0,x) # 0 과 x 중 큰값을 리턴하라는 것 
```

```python
y_dash = ReLU(x)
plt.plot(x, y_dash)
plt.ylim(-0.1, 5.1)
plt.show()
```
