```python
import numpy as np
import matplotlib.pyplot as plt  #pyplot 모듈을 이용해봅시다 
```

```python
x = np.arange(0, 6 , 0.1) #0 부터 6까지 0.1 간격으로 생성 
y = np.sin(x)
```

```python
#그래프 그리기
plt.plot(x,y)
```

```python
plt.show()
```

이제 추가로 COS 도 그려보겠습니다 


```python
x = np.arange(0,10, 0.1)
```

```python
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
plt.plot(x, y1, label='sin')
plt.plot(x, y2, label='cos', linestyle="--") #cos 함수는 점선으로 그리기 
plt.xlabel("x") #x축 이름 
plt.ylabel("y") #y축 이름
plt.title('sin&cos')#제목
plt.legend()
plt.show()
```
